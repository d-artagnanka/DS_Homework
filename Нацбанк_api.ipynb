```python
# https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?valcode=USD&date=20201202&json
import requests
import json
import pickle 

r = requests.get("https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?valcode=USD&date=20201202&json")
_date = json.loads(r.content.decode("utf-8"))[0]
print(_date)
rate, exdate = _date['rate'], _date['exchangedate']
print(exdate, rate)
```

```python
out_d = {}

for mn in range(1, 13):
    daily = []
    for day in range(1, 32):
        url = 'https://bank.gov.ua/NBUStatService/v1/statdirectory/'\
              'exchange?valcode=USD&date=2020{m}{d}&json'.format(m=str(mn).zfill(2), d=str(day).zfill(2))
        resp = requests.get(url)
        
        try:
            _date = json.loads(resp.content.decode("utf-8"))[0]
            if 'message' in _date.keys():
                continue
            else:
                daily.append(_date['rate'])
        except IndexError:
            break
    
    out_d[mn] = daily
    
print(out_d)
```

```python
middle_value = []
for i in out_d.keys():
    sum = 0.0
    for day in out_d[i]:
        sum += day
    middle_value.append(round(sum/len(out_d[i]), 4))

middle_value[]
```

```python
import math
deviation = []
for i in out_d.keys():
    dev =0.0
    for day in out_d[i]:
        dev += (day - middle_value[i-1])**2
    deviation.append(math.sqrt(dev/len(out_d[i])))
```

```python
total=[]
for i in range(0, 12):
    total.append([middle_value[i], deviation[i]])
```

```python
with open('result_file.pickle', 'wb') as f:
    pickle.dump(total, f)
```
